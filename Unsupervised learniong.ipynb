# Unsupervised learning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#clustering
from sklearn.cluster import AgglomerativeClustering, KMeans #clustering mean unsupervised learning
```

```python
data = pd.read_csv('../dataset for Python/Mall_Customers.csv')
```

```python
X= data[['Annual Income (k)','Spending Score (1-100)']] #this is the sencond method to retrive the column
X
```

```python
plt.scatter(X['Annual Income (k)'], X['Spending Score (1-100)'])
plt.show()
```

```python
sse = []
for clusters in range(1,10):
    k = KMeans(n_clusters=clusters)
    k.fit(X)
    sse.append(k.inertia_)
```

```python
plt.plot(range(1,10), sse)
plt.xlabel('Cluster(k)')
plt.ylabel('Sum of Squared Error')
plt.title('cluster')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=666)
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
cluster_labels = kmeans.labels_
```

```python
X=X.values
```

```python
plt.scatter(X[cluster_labels==0,0], X[cluster_labels==0,1], c='r')
plt.scatter(X[cluster_labels==1,0], X[cluster_labels==1,1], c='b')
plt.scatter(X[cluster_labels==2,0], X[cluster_labels==2,1], c='g')
plt.scatter(X[cluster_labels==3,0], X[cluster_labels==3,1], c='y')
plt.scatter(X[cluster_labels==4,0], X[cluster_labels==4,1], c='c')
plt.xlabel('Annual income')
plt.ylabel('Spending Score')
plt.show()
```

```python
agg = AgglomerativeClustering(n_clusters=5)
agg.fit(X)
cluster_labels= agg.labels_
```

```python
plt.scatter(X[cluster_labels==0,0], X[cluster_labels==0,1], c='r')
plt.scatter(X[cluster_labels==1,0], X[cluster_labels==1,1], c='b')
plt.scatter(X[cluster_labels==2,0], X[cluster_labels==2,1], c='g')
plt.scatter(X[cluster_labels==3,0], X[cluster_labels==3,1], c='y')
plt.scatter(X[cluster_labels==4,0], X[cluster_labels==4,1], c='c')
plt.xlabel('Annual income')
plt.ylabel('Spending Score')
plt.show()
```
