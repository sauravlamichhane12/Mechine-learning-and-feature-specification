```python
norm = np.random.normal(6,9,1000)
```

```python
class StandardScaler:
    def __init__(self,):
        self.mean=None
        self.std=None
    def fit(self,data):
        self.mean=data.mean()
        self.std=data.std()
    def transform(self, data):
        return(data - self.mean)/self.std
```

```python
scaler = StandardScaler()
scaler.fit(norm)
```

```python
scaled_norm=scaler.transform(norm)
```

```python
norm
```

```python
scaled_norm
```
