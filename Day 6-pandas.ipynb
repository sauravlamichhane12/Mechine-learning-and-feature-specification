```python
y3 = y1 **3
```

```python
y3
```

```python
plt.figure(figsize=(20,10))
plt.title('My plot')
plt.xlabel('Speed')
plt.ylabel('Distance')
plt.plot(x, y1, color='r', marker='X', label='Red')
plt.plot(x,y2, color='b', marker='*', label='Blue')
plt.plot(x,y3, color = 'y', marker ='+', label ='yellow')
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```
